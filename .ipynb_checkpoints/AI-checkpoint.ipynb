```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

# Créer une instance de la classe LinearRegression
regressor = LinearRegression()

# Fitter le modèle à nos données d'entrée
regressor.fit(x_train, y_train)

# Afficher les coefficients de régression et l'ordonnée à l'origine
print("Coefficients de régression : ", regressor.coef_)
print("Ordonnée à l'origine : ", regressor.intercept_)

# Afficher l'expression du modèle de la régression linéaire
print("Modèle de régression linéaire : y = ", regressor.intercept_, "+", 
      regressor.coef_[0], "* x1 +", regressor.coef_[1], "* x2 +", regressor.coef_[2], "* x3 + ...")
```

```python
y_test_predict=regressor.predict(x_test)# Use the trained linear regression model to make predictions on the test data
#q17
```

```python
# Comparaison des valeurs prédites aux valeurs réelles# Create a DataFrame to compare the predicted and actual values

df_pred_vs_actual = pd.DataFrame({'Valeurs réelles': y_test_predict, 'Valeurs prédites': y_test_predict})
print(df_pred_vs_actual)# Output the DataFrame
```

```python
# Use the trained linear regression model to make predictions on the test data
y_test_predict = regressor.predict(x_train)

# Create a DataFrame to compare the predicted and actual values
df_train_pred_vs_actual = pd.DataFrame({'Valeurs réelles': y_train, 'Valeurs prédites': y_test_predict})

# Output the DataFrame
df_train_pred_vs_actual
```

```python

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Prédiction des valeurs de profit à partir des données de test
y_pred = regressor.predict(x_test)

# Calculer MSE
mse = mean_squared_error(y_test, y_pred)
print('MSE:', mse)

# Calculer MAE
mae = mean_absolute_error(y_test, y_pred)
print('MAE:', mae)

# Calculer RMSE
rmse = np.sqrt(mse)
print('RMSE:', rmse)

# Calculer le coefficient R2
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```
